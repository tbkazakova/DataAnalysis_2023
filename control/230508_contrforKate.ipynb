# Игра в слова
Гоша вспоминает слова на одну букву. Помогите ему.

Сначала примите от пользователя букву.

Теперь принимайте слова, пока он не введёт `ВСЁ`.

Если слово начинается не на нужную букву, выведите `Это на другую букву.`

После ввода "ВСЁ" выведите уникальные слова, которые он вспомнил и их количество.

Пример:

```
а
азбука
арбуз
автобус
ааа
автобус
облако
Программа выводит: "Это на другую букву".
ананас
ВСЁ
Программа выводит: "азбука, арбуз, автобус, ааа, ананас"
Программа выводит: "Всёго: 5"
```

````python
a = str(input())
c = ''
d = 0
while True:
    b = str(input())
    if b == 'ВСЁ':
        break
    if b[0] != a:
        print("Это на другую букву")
    else:
        c = c + b + ', '
        d += 1
print(c[:-2])
print('Всёго: ', d)
````
