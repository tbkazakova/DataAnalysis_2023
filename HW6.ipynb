Порядок слов в русском языке часто имеет специальный смысл. Под порядком слов подразумевается, прежде всего, порядок сказуемого (обозначим его *V*, от verb), подлежащего (обозначим его *S*, от subject) и объекта (*O*, от object).

Кроме указанных членов предложения, можем также встретить наречия (*Adv* &mdash; adverbials), предлоги (*Adp* — adpositions), какие-то существительные (кроме подлежащего и объекта; *Noun*) и прилагательные (*Adj* — adjectives).

Порядок **V S (O)** часто встречается в анекдотах. Предложение *Заходит однажды улитка в бар*, которое выглядит в нашей записи как `V Adv S Adp Noun` можно считать **анекдотным**. Эти предложения определяются тем, что сказуемое расположено до подлежащего.

С другой стороны, порядок **O S V** часто встречается при ответе на вопрос `кого? что?` и на некоторые другие вопросы, при исправлении, или когда объект O — местоимение или просто фонетически короче, чем подлежащее S. Такие предложения можно называть **нетривиальными**. Например, предложение *Ефросинью Евграф сильно любит* (`O S Adv V` в нашей записи) нетривиальное.

Пользователь вводит порядок слов в предложении. В предложении всегда есть V (сказуемое). В предложении могут быть, а могут не быть, S и O. Кроме того, там свободно могут быть другие слова, перечисленные выше (*Adv*, *Adp*, *Noun*, *Adj*).
- Если порядок слов — **анекдотный**, то выведите `Хорошее начало анекдота. Что там дальше?`.
- Если порядок слов — **нетривиальный**, выведите `Интересно сказано!` 
- Иначе выведите, что `Всё понятно.` 


**Бонус (+1 балл):** вы учли, что в анекдотном порядке до сказуемого не должно быть никаких имён (`Noun`). Т.е. Noun, это существительные кроме O или S, и их не должно быть до V S для анекдотного порядка. 

Примеры:

Ввод, вывод:
    
    V Adv S Adp Noun
    Хорошее начало анекдота. Что там дальше?

Ввод, вывод:
    
    V Adv Noun S
    Хорошее начало анекдота. Что там дальше?
    
Ввод, вывод (**бонус за `"всё понятно"`**, нет бонуса за `"Хорошее начало анекдота. Что там дальше?"`):  
    
    Noun V S Adv
    Всё понятно.

Ввод, вывод:

    O S Adv V
    Интересно сказано!

Ввод, вывод:

    O Adp Noun S V
    Интересно сказано!

Ввод, вывод:

    V
    Всё понятно.

Организаторы награждают спортсменов.

- Спортсмен, занявший последнее место, награждается со словами `"<имя> награждается призом зрительских симпатий!"`.
- Остальные, занявшие место ниже третьего, награждаются со словами `"Награждается <имя> за <само место (4, 5, 12 и т.д.)> место!"`;
- Первые три человека награждаются особо, со словами `"Особо награждается <имя> за почётное <само место (1, 2, 3)> место!"`;
- Организаторы ошиблись с настройками табло, и поэтому после каждого награждения программа также выводит (в строчку через пробел) список спортсменов, пока остающихся без наград, наградить которых ещё предстоит.
