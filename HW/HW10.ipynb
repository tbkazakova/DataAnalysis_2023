**Подсказка-пример №1:** 

````python
age = np.array([4, 20, 45, 80])  # это возрасты людей

# У кого скидка в музей? Просто массив True/False.
print((age < 18) | (age > 75))

# Можно наложить на массив такой же длины
# и вывести элементы с индексами, соответствующими True

print(age[(age < 18) | (age > 75)])

# имена - массив такой же длины
names = np.array(["Стёпка", "Вера", "Надежда", "Любовь"])

# тогда, имена тех, кто может взять скидку:
print(names[(age < 18) | (age > 75)])
````

**Подсказка-пример №2:** фильтрация и обращение по индексам - устроены одинаково. Как и в примере выше, так и в многомерном массиве можно вывести информацию по индексу, по интервалу индексов, а можно по условию про массив с таким же количеством элементов. (см. пример)

````python
athletes = np.array([
    # номер спортсмена, рост, вес, жим от груди
    [0, 172, 70, 65],
    [1, 181, 70, 80],
    [2, 178, 72, 71],
    [3, 190, 100, 160]])

print('Вес:', athletes[:, 2])  # все строки, второй столбец

# полная информация о спортсменах, жмущих больше собственного веса
print('\nСилачи:')
print(athletes[athletes[:, 2] < athletes[:, 3], :])  # строки, где вес меньше жима, все столбцы
````
