# О проекте

Курс устроен так, что мы довольно мало успели обсудить из того, чем программирование и анализ данных помогают филологии. Небольшой проект поможет чуть больше про это узнать.

**Идея:** в группах придумать и выполнить небольшое задание (исследование?), которое имеет значимость (пусть небольшую) в филологии и при этом упрощается использованием питона (/экселя)

### Кратко основная информация:
- группы: `2-5 человек` (в крайнем случае можно одному, но согласуйте с семинаристом)
- результат: устный рассказ с презентацией + код и пояснения к нему на гитхабе
- согласование темы: до `8 июня`, лучше раньше ([таблица](https://docs.google.com/spreadsheets/d/1vyQ2D1VYG-dX67MuIeKOeErE5zVGn-kryUo0IV4n7r0/edit?usp=sharing))
- защиты: `23 июня` (очно, но при особых обстоятельствах можно в зуме)

### Подробнее о темах и ожидаемых результатах:

В качестве темы принимается что угодно, если вы можете обосновать, что это полезно филологии, и при этом вы или семинаристы понима(ете|ют), как реализовать такое на питоне.

Примеры тем (можно модифицировать эти, можно брать совершенно другие):
- **анализ текста с применением визуализаций**
- в том числе улучшения игрушечного примера с **граф**ом (вариации: когда лучше проводить ребро между вершинами, что делать вершинами (можно места, например, а не персонажей), "схлопывание" разных имён одного персонажа, цвета-подписи и т.д.)
- **стилометрия**, сравнение похожести текстов и авторов в целом
- **проверка стереотипа или интересной гипотезы** (пример: правда ли, что у Л.Н. Толстого длинные предложения)
- супертема: **найти датасет** (как таблица с велосипедами, но про филологическое), на котором можно проиллюстрировать основные темы курса по анализу данных, и **показать** реалистичные (не притянутые за уши) примеры этого. Темы: сортировка, фильтрация, группировка, описательная статистика: среднее, медиана, мода, квартили, дисперсия, стандартное отклонение; выбросы, заполнение пропущенных значений, корреляция, регрессия. Вам могут помочь сайты https://www.kaggle.com/datasets, https://dataverse.pushdom.ru/ и другие.
- ...

До 8ого июня занесите в [таблицу](https://docs.google.com/spreadsheets/d/1vyQ2D1VYG-dX67MuIeKOeErE5zVGn-kryUo0IV4n7r0/edit?usp=sharing) участников группы, вашу тему и краткое пояснение, что вы собираетесь делать и какими инструментами. Мы их посмотрим и утвердим / порекомендуем инструменты / попросим пояснить/изменить.

Это нужно, чтобы вы были уверены, что занимаетесь тем, что на защитах оценят. Обсуждать с Максимом, Таней, Антоном и Олегом можно и нужно. Написать семинаристу даже "мы не знаем, что делать, но любим советские фильмы" гораздо лучше, чем ничего. Мы что-нибудь подскажем, посоветуем.

На протяжении работы тоже можно спрашивать. Лучше мы сделаем проект с вами, чем его не сделает никто.

### Про защиту и оценку

Вас будет около 28 групп... И мы всех послушаем

**На защиту:** 5 минут устного рассказа: что смотрели, как смотрели, какие выводы. Потом ответы на вопросы. На **гитхабе** одного из участников должен лежать **код с пояснениями**. Ссылку на репозиторий и на презентацию надо будет положить в [таблицу](https://docs.google.com/spreadsheets/d/1vyQ2D1VYG-dX67MuIeKOeErE5zVGn-kryUo0IV4n7r0/edit?usp=sharing).

С вопросами и переходами на один доклад будет уходить около 10 минут.

Сделаем так: 6 защит (час) - перерыв (15 минут) - 6 защит - перерыв ...

Защиты очные, в крайнем случае можно в зуме. Вы распределитесь по слотам, чтоб не сидеть весь день.

**Оцениваться** будет наличие "исследовательского вопроса" или задачи и "ответа"-решения, разумность подхода, объём, сложность и ценность сделанного, новизна относительно курса и мира, наличие понятных комментариев к коду для его применимости в подобных исследованиях.

В целом: этим проектом мы хотим поднять вам оценку, а не сделать хуже.

### Вопросы и ответы:

- Можно ли анализировать всё в **экселе**, а питон не трогать?
    - Если ваша задача и вправду не выигрывает от использования питона - да, можно! Но такие проекты будут строже оцениваться с точки зрения разумности подхода, объёма сделанного и ценности результатов. Например, для проверки того, что у Толстого длинные предложения, питон необязателен, но от вас требуется понимание и использование статистических методов (например, статистической значимости).


- Можно объединяться людям из разных групп? Из 211 и 214, например.
    - Конечно!


- А что если две группы будут делать одно и то же?
    - Совсем одно и то же делать нельзя. Но такого не произойдёт, ведь темы групп согласовываются заранее. Возможно, некоторые будут похожи, мы не так много прошли, это нормально.
